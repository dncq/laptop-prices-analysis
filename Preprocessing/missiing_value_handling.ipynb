# Handling CPU

```python
def handle_duplicate(txt):
    try:
        new_word = []
        for word in txt.strip().replace("-",' ').split():
            if word not in new_word:
                if word.capitalize() == "Family" or word.capitalize() == "Processor":
                    continue
                if word == 'core' and word.capitalize() in new_word:
                    continue
                new_word.append(word)
        return " ".join(w for w in new_word)
    except:
        return None
```

```python
for i in range(df.shape[0]):
    df.at[i, 'CPU'] = handle_duplicate(df.at[i,'CPU'])

df
```

```python
missing_cpu_indices = df.index[df['CPU'].isna()].tolist()
for i in missing_cpu_indices:
    df = df.drop(i)

df.info()
```
