```python
def get_spec(url):
    spec = {}
    spec['Brand'] = ""
    spec['Series'] = ""
    spec['Screen Size'] = ""
    spec['Screen Resolution'] = ""
    spec['CPU Manufacturer'] = ""
    spec['CPU Model'] = ""
    spec['CPU Speed'] = ""
    spec['RAM'] = ""
    spec['GPU'] = ""
    spec['GPU Type'] = ""
    spec['Storage'] = ""
    spec['Storage Type'] = ""
    spec['Power'] = ""
    spec['Weight'] = ""
    spec['Price'] = ""
    
        
    browser.get(url)

    soup2 = BeautifulSoup(browser.page_source, 'html.parser')
    
    try:
        price = soup2.find_all("span",{"class": "a-price aok-align-center"})
        spec['Price'] = float(price[0].text.replace('$',' ').split().pop())
        sleep(1)
    except:
        spec['Price'] = ""
    for tr in soup2.findAll("tr"):
        name = tr.find('th', {'class': 'a-color-secondary a-size-base prodDetSectionEntry'})
        if name != None:
            if name.text.strip() == "Brand":
                brand = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['Brand'] = brand
            if name.text.strip() == "Series":
                series = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['Series'] = series
            if name.text.strip() == "Screen Size":
                ss = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['Screen Size'] = ss
            if name.text.strip() == "Screen Resolution":
                sr = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['Screen Resolution'] = sr
            if name.text.strip() == "CPU Manufacturer":
                cpum = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['CPU Manufacturer'] = cpum
            if name.text.strip() == "CPU Model":
                cpu_model = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['CPU Model'] = cpu_model
            if name.text.strip() == "CPU Speed":
                cpus = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['CPU Speed'] = cpus
            if name.text.strip() == 'Ram Memory Installed Size':
                ram = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['RAM'] = ram
            if name.text.strip() == "Graphics Coprocessor":
                gpu = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['GPU'] = gpu
            if name.text.strip() == "Card Description":
                gpu_type = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['GPU Type'] = gpu_type
            if name.text.strip() == "Hard Disk Size":
                hds = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['Storage'] = hds
            if name.text.strip() == "Hard Disk Description":
                hdd = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['Storage Type'] = hdd
            if name.text.strip() == "Voltage":
                vol = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['Power'] = vol
            if name.text.strip() == "Item Weight":
                iw = tr.find('td', {'class': 'a-size-base prodDetAttrValue'}).text.strip()
                spec['Weight'] = iw         
    return spec
        
    
```

```python
url_list = []
data = defaultdict(list)
for i in range(1,21):
    page_url = "https://www.amazon.com/s?k=Laptop+Computer&page={}&crid=761XKW8RXAQ5&qid=1701177074&sprefix=laptop+c%2Caps%2C775&ref=sr_pg_{}".format(i,i)

    browser.get(page_url)
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    print("Scraping page", str(i))
    
    results = soup.find_all("h2", {"class": "a-size-mini a-spacing-none a-color-base s-line-clamp-2"})
    for item in results:
        url_list.append(parent_url + item.find('a').get('href'))
print(len(url_list))    
```
