## Saving Name

```python
import json
import os


for i in range(100):
# Path to the saved JSON file
    filename = f"assets/new_egg/scraped_json/page{i+1}_data.json"
    output_file = f"assets/new_egg/extracted_name/page{i+1}.txt"
    print(f"Currently at page {i}")
    try:

        with open(filename, 'r') as file:
            loaded_data = json.load(file)

        # Access the loaded data as needed
        product_dict = loaded_data["Products"]


        # Open the output file in write mode
        with open(output_file, 'w') as output_file:
            # Iterate through each product and concatenate BulletDescriptions into a single line
            for j in range(96):
                print(f"Currently at product{j} page {i}")
                if product_dict[j]['ItemCell']['Description']['Title'] is not None:
                    title = product_dict[j]['ItemCell']['Description']['Title']
                    title_single_line = ' '.join(title.splitlines())
                    output_file.write(f"{title_single_line}\n")
                else:
                    print(f"Encountering NOne Name at page {i+1}, product{j+1})")
                    output_file.write(f"None\n")

    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
    except FileNotFoundError:
        print(f"File not found: {filename}")
```

## Saving BulletDescription

```python
import json
import os


for i in range(100):
# Path to the saved JSON file
    filename = f"assets/new_egg/scraped_json/page{i+1}_data.json"
    output_file = f"assets/new_egg/extracted_BD/page{i+1}.txt"
    print(f"Currently at page {i}")
    try:

        with open(filename, 'r') as file:
            loaded_data = json.load(file)

        # Access the loaded data as needed
        product_dict = loaded_data["Products"]


        # Open the output file in write mode
        with open(output_file, 'w') as output_file:
            # Iterate through each product and concatenate BulletDescriptions into a single line
            for j in range(96):
                print(f"Currently at product{j} page {i}")
                if product_dict[j]['ItemCell']['Description']['BulletDescription'] is not None:
                    bd = product_dict[j]['ItemCell']['Description']['BulletDescription']
                    bd_single_line = ' '.join(bd.splitlines())
                    output_file.write(f"{bd_single_line}\n")
                else:
                    output_file.write(f"None\n")

    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
    except FileNotFoundError:
        print(f"File not found: {filename}")
```

## Saving Final Price

```python
import json
import os


for i in range(100):
# Path to the saved JSON file
    filename = f"assets/new_egg/scraped_json/page{i+1}_data.json"
    output_file = f"assets/new_egg/extracted_fp/page{i+1}.txt"
    print(f"Currently at page {i+1}")
    try:

        with open(filename, 'r') as file:
            loaded_data = json.load(file)

        # Access the loaded data as needed
        product_dict = loaded_data["Products"]


        # Open the output file in write mode
        with open(output_file, 'w') as output_file:

            # Iterate through each product and concatenate BulletDescriptions into a single line
            for j in range(96):
                print(f"Currently at product{j} page {i}")
                if product_dict[j]['ItemCell']['FinalPrice'] is not None:
                    price = product_dict[j]['ItemCell']['FinalPrice']

                    output_file.write(f"{price}\n")
                else:
                    output_file.write(f"None\n")

    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
    except FileNotFoundError:
        print(f"File not found: {filename}")
```
